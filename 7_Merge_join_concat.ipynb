```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
# Concat =>  concatenacao, juncao dos dados
# juntar os dados (abre uma lista) junta todos df, vai encaixar um abaixo do outr

pd.concat([df1, df2, df3])
```

```python
# concatenar horizontalmente, ele espera que tenhamos indices iguais
# ele expande linhas, porque tem indices diferentes
# ele espera indices iguais
pd.concat([df1, df2, df3], axis=1) # ele vai expandir horizontalmente
```

```python
## Outros DataFrames  - Merge
```

```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
```

```python
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})  
```

```python
esquerda
```

```python
direita
```

```python
# merge => coluna chave semelhante, com colunas diferentes
# # method que encaixa os dois
pd.merge(esquerda, direita)
```

```python
# mesclar e manter os dados
pd.merge(esquerda, direita, on="key") # => preserva a chave key
```

```python
# outro exemplo com key diferentes
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
# fazer a selecao em duas colunas no mesmo tempo
pd.merge(esquerda, direita, on=['key1', 'key2']) # ele mescla e mantem as outras linhas
# ele ve os valores similares => O inner join retorna apenas as linhas em que as chaves key1 e key2 estão presentes nos dois DataFrames (ou seja, que coincidem nos dois lados).
# Apenas linhas com chaves comuns nos dois DataFrames
# Todas as linhas de ambos os DataFrames (com NaN onde faltar)
```

```python
# parametro how (como eu quero que isso seja feito)
pd.merge(esquerda, direita, how="outer", on=['key1', 'key2'])
# Ele mantém todas as linhas de ambos os DataFrames.
# Onde não houver correspondência entre as chaves (key1 e key2), 
# os valores ausentes serão preenchidos com NaN.
```

```python
# qdo nao encaixar os dados coloca o NaN
pd.merge(esquerda, direita, how="right", on=['key1', 'key2'])
```

```python
# qdo nao encaixer coloca na esquerda
pd.merge(esquerda, direita, how="left", on=['key1', 'key2'])
# Mantém todas as linhas do DataFrame da esquerda (esquerda), e só traz os dados do direita quando houver correspondência nas chaves.
# Se não tiver correspondência, ele preenche os dados do direita com NaN.
```
